### Из теории.
<big>
Задача нахождения безусловного максимума функции $f(x)$ векторного аргумента $x={x_1,x_2,...,x_n}$, заданной на всём пространстве $R^n$, символически записываемую в виде: 
<br />
<br />
    <center> $f(x)\to max, \ x\in R^n$. (1)</center>
<br />

Точка $\bar{x}$ называется точкой локального максимума в задаче (1), если существует число $\epsilon > 0$ такое, что для всех $x$ удовлетворяющих условию $\parallel x - \bar{x}\parallel\le\epsilon$, выполнено неравенство $f(x)\le f(\bar{x})$; здесь и далее $\parallel x\parallel=\{ x_1^2+...+x_n^2\}^{1/2}$. Точка $\bar{x}$ называется точкой глобального максимума в задаче (1), если $f(\bar{x})\ge f(x), \ x\in R^n$.
Для численного решения задачи (1) обычно строят некоторую последовательность векторов $\{ x^k\}_{k=0}^{\infty}$, обрывая процесс построения тогда, когда появляется уверенность, что последовательность, что последний из построенных элементов последовательности близок к точке максимума в том или ином смысле. Например,если элемент $x^k$ удовлетворяет неравенству $\parallel x^k-\bar{x}\parallel <\bar{\delta}$, где $\bar{x}$ - точка максимума в задаче (1), $\bar{\delta}$ - некоторое положительное число, то $x^k$ есть приближенное решение задачи (1) с точностью $\bar{\delta}$ по аргументу. Если же выполняется неравенство $\mid f(x^k)-\bar{f}\mid < \bar{\delta}$, где $\bar{f}$ - максимальное значение функции $f(x)$ в $R^n$, то число $f(x^k)$ даёт приближенное решение задачи (1) с точностью $\bar{\delta}$ по функции.
Большинство процессов, используемых для приближенного решения задачи (1) можно представить как итерационные в виде:
<br />
<br />
<center>$x^{k+1}=x^k+\alpha_kp^k$,  (2)</center>
<br />
где $p^k$ - вектор определяющий направление движения от точки $x^k$ к точке $x^{k+1}$, $\alpha_k$ - числовой множитель, величина которого задаёт длину шага в направлении $p^k$. В различных итерационных процессах типа (2) для нахождения $\alpha_k$ и $p^k$ привлекаются различные сведения о минимизируемой функции $f(x)$. Процессы, использующие для этого только значения самой функции, называются процессами нулевого порядка или методами поиска. Процессы, требующие вычисления производных $f(x)$ до $m$-го порядка включительно называют процессами $m$-го порядка. Если для определения $\alpha_k$ и $p^k$ используется информация, полученная на S предыдущих итерациях, то итерационный процесс называется $S$-шаговым.
В качестве направления $p^k$ естественно выбирать направление возрастания функции $f(x)$. $p^k$ при нахождении максимума - направление подъема, а число $\alpha_k$ таково:
<br />
<center>$f(x^{k+1})=f(x^k+\alpha_kp^k)>f(x^k)$. (3)</center>
<br />
Важной характеристикой итерационного процесса является скорость сходимости итерационной последовательности. Пусть итерационная последовательность $\{x^k\}_{k=0}^\infty$ сходится к $\bar{x}$, т.е.:
<br /> 
<br />
<center>$\parallel x^k - \bar{x}\parallel\to 0, \ \ k\to\infty$.</center>
<br />
Скорость сходимости последовательности 
<br> 1) линейна, если существует число $q\in(0,1)$ такое, что 
<br />
<center>$\parallel x^{k+1}-\bar{x}\parallel\le q\parallel x^k-\bar{x}\parallel, \ k=0,1,2,...;$</center>
<br /> 2) сверхлинейна, если существует последовательность чисел $\{q_k\}_{k=1}^\infty, q_k\to 0, k\to\infty$ такая, что
<br />
<center>$\parallel x^{k+1}-\bar{x}\parallel\le q_k\parallel x^k-\bar{x}\parallel, \ k=0,1,2,...;$</center>
<br /> 3) квадратична, если существует постоянная $C > 0$ такая, что 
<br />
<center>$\parallel x^{k+1}-\bar{x}\parallel\le C\parallel x^k-\bar{x}\parallel^2, \ k=0,1,2,...;$</center>
</big>

<big>
<br /> 
**Метод сопряженных градиентов**
<br>
В этом методе $\alpha_k$ выбирается из условия $f(x^k+\alpha_kp^k)=\underset{\alpha>0}{max}\ f(x^k+\alpha p^k)$. 
  <br>  При этом вектор $p^k$ зависит не только от градиента функции $f(x^k)$, но и от градиента в предыдущей точке $\bigtriangledown f(x^{k+1})$, и строится по правилу:
    <br />
    <center> A) $p^k=-\bigtriangledown f(x^k) + \beta_kp^{k-1}$, </center>
    <br><center>$\beta_0=0,\beta+k=\frac{(\bigtriangledown f(x^k),\bigtriangledown f(x^k)-\bigtriangledown f(x^{k-1)}))}{\parallel\bigtriangledown f(x^{k-1})\parallel^2}$, $k = 1,2,...;$</center>
    <br> либо по правилу
    <br />
    <br />
    <center>Б) $p^k=-\bigtriangledown f(x^k)+\beta_kp^{k-1}$,</center>
    <br />
    <center>$\beta_k=\begin{cases}{0,k=0,n,2n,...;  \\}{\frac{(\bigtriangledown f(x^k),\bigtriangledown f(x^k)-\bigtriangledown f(x^{k-1)}))}{\parallel\bigtriangledown f(x^{k-1})\parallel^2}, \ k = 1,2,...; \\ }{k=0,...,n-1,n+1,...,2n-1,2n+1,...}\end{cases}$</center>
    <br /> 
</big>

<big>$n$ - размерность пространства независимых переменных.</big>

<big>**Эврестический алгоритм Свенна нахождения интервала унимодальности.**
<br> Для выбранной исходной точки $x^0$ и заданного $h$ на основе правила исключения строится относительно широкий интервал, содержащий точку оптиума. Используется эврестический подход, в котором $x^{k+1}$ пробная точка определяется по рекурентной формуле:
<br />
    <br />
<center>$x^{k+1}=x^k+\frac{h}{2}, \ \ \ k=0,1,2,...$,</center>
где 
    <br>$x^0$ - произвольно выбранная начальная точка;
    <br>$h$ - шаг поиска, знак которого может меняться на противоположный.
    <br>Знак $h$ определяется путем сравнения значений $f(x^0),f(x^0+\mid h\mid)$ и $f(x^0-\mid h\mid)$. Если $f(x^0-\mid h\mid)\ge f(x^0)\ge f(x^0+\mid h\mid)$, то величина $h$ выбирается отрицательной. Если $f(x^0-\mid h\mid)\le f(x^0)\le f(x^0+\mid h\mid)$, то величину $h$ следует выбирать положительной. Если $f(x^0-\mid h\mid)\le f(x^0)\ge f(x^0+\mid h\mid)$, то точка максимума лежит между $x^0-\mid h\mid$ и $x^0+\mid h\mid$ и поиск граничных точек завершен, в противном случае изменить начальную точку. Случай, когда $f(x^0-\mid h\mid)\ge f(x^0)\le f(x^0+\mid h\mid)$, противоречит предположению об унимодальности. Выполнение этого условия говорит о том, что функция в окрестности точки $x^0$ не является унимодальной и следует изменить начальную точку.
</big>

<big>
    **Метод золотого сечения.**
<br> Поиск с помощью метода золотого сечения основан на разбиении отрезка неопределенности на две части, известном как "золотое сечение". При этом отношение длины всего отрезка к большей части равно отношению большей части к меньшей и равно числу $\tau=2^{-1}(1+\sqrt{5})\simeq1.6118$ ($\tau$ - корень уравнения $\tau^2=1+\tau$). В методе золотого сечения точки $x^1$ и $x^2$ на каждом отрезке неопределенности $[a,b]$ выбираются оп правилу:
<br />
<center>$x^1=b-(b-a)/\tau, x^2=a+(b-a)/\tau$.</center>
<br />
Показатель эффективности метода равен $E_m=\tau^{1-m}$.
    </big>

```julia
#КОД.

#Выбираем начальные значения переменных.
x1=0
x2=0
x3=0

y1=0
y2=0
y3=0

z1=0
z2=0
z3=0

px1=0
px2=0
px3=0

o=0
x11=0
x22=0
alpha=0

h=0.001 
E=0.001 
k=0 
N=3 
a=zeros(10000) 
a[1]=0 

n=1
A=0
B=0 
p=0 
o=0 
xx1=zeros(23)
xx2=zeros(23)
xx3=zeros(23)
xx1[1]=0
xx2[1]=0
xx3[1]=0
```

```julia
#Функции.
f(x1,x2,x3)=4-2*(x1.-3).^2-(x2.-2).^2-3*(x3.+1).^4 #Исходная функция.
gradf(x1,x2,x3)=-4*(x1.-3).-2*(x2.-2).-12*(x3.+1).^3 #Градиент исходной функции.

gradf1(x1)=-4*(x1-3) #Производная по x1.
gradf2(x2)=-2*(x2-2) #Производная по x2.
gradf3(x3)=-12*(x3+1)^3 #
```

```julia
#Выражаем x1,x2,x3 через alpha (здесь - a) и подставляем в функции.
ff(a)=4-2*(x1+a*px1-3)^2-(x2+a*px2-2)^2-3*(x3+a*px3+1)^4
gradff(a)=-4*(x1+a*px1-3)-2*(x2+a*px2-2)-12*(x3+a*px3+1)^3
gradff1(a)=-4*(x1+a*px1-3)
gradff2(a)=-2*(x2+a*px2-2)
gradff3(a)=-12*(x3+a*px3+1)^3
```

```julia
# Метод сопряженных градиентов.
while((abs(gradf1(x1))+abs(gradf2(x2))+abs(gradf3(x3)))>E)
if(k==0)
px1=-gradf1(x1)
px2=-gradf2(x2)  
px3=-gradf3(x3)
end

#Реализуем алгоритм Свенна для поиска интервала унимодальности
while(true)
    if((ff(a[n]-h)<=ff(a[n]))&&(ff(a[n])<=ff(a[n]+h)))
        a[n+1]=a[n]+h
    elseif((ff(a[n]-h)>=ff(a[n]))&&(ff(a[n])>=ff(a[n]+h)))
        a[n+1]=a[n]-h
        h=-h
    elseif((ff(a[n]-h)<=ff(a[n]))&&(ff(a[n])>=ff(a[n]+h)))
        A=a[n]-h
        B=a[n]+h
        break
end

n+=1
a[n+1]=a[n]+h
    
while(ff(a[n+1])>=ff(a[n]))
    n+=1
    a[n+1]=a[n]+h  
end
    
if(h>0)
    A=a[n-1]
    B=a[n+1]
else 
    A=a[n+1]
    B=a[n-1]
end
break
end
n=1

# Определяем альфу из золотого сечения
t=(1+sqrt(5))/2
x11=B-(B-A)/t
x22=A+(B-A)/t
alpha=(x11+x22)/2

# Продолжаем алгоритм метода сопряженных градиентов
# Ищем новые "иксы".
y1=x1
x1=y1+alpha*px1
xx1[o+2]=x1

y2=x2
x2=y2+alpha*px2
xx2[o+2]=x2

y3=x3
x3=y3+alpha*px3
xx3[o+2]=x3

o+=1    
if((abs(gradf1(x1))+abs(gradf2(x2))+abs(gradf3(x3)))<E)
    break
end
    
if((k+1)==N)
    k=0
    continue
end

#Ищем новое направление pk.    
betta1=(16*((x1-3)-(y1-3))*(x1-3))/(16*(y1-3)^2)
px1=-gradf1(x1)+betta1*px1

betta2=(4*((x2-2)+(y2-2))*(x2-2))/(4*(y2-2)^2) 
px2=-gradf2(x2)+betta2*px2

betta3=(144*((x3+1)^3+(y3+1)^3)*(x3+1)^3)/(144*(y3+1)^6)
px3=-gradf3(x3)+betta3*px3
   
k+=1
    
println("x1 =  ", x1)
println("x2 =  ", x2)
println("x3 =  ", x3)
println("\n")
    
end
println("Количество итераций: ", o+1)
```

## Метод Нелдера-Мида.

```julia
ff(x)=-(4-2*(x[1]-3).^2-(x[2]-2).^2-3*(x[3]+1).^4)
x1=[0.,1.,-1.]
k=-0.5
e1=[1,0,0]
e2=[0,1,0]
e3=[0,0,1]
x2=x1+k*e1
x3=x1+k*e2
x4=x1+k*e3
b=[]
g=[]
w=[]
mid=[]
xr=[]
a=1
xe=[]
y=2
xr1=[]
xs=[]
B=0.5
n=-1
b=x1
w=x2
g=x3
e=0.0001

q=zeros(4)

q[1]=ff(x1)
q[2]=ff(x2)
q[3]=ff(x3)
q[4]=ff(x4)

r=copy(q)

t=0

sort!(q)

for i=1:4
    if(q[1]==r[i])
        t=i
        if(t==1)
            b=x1
        end
        if(t==2)
            b=x2
        end
        if(t==3)
            b=x3
        end
        if(t==4)
            b=x4
        end
    end
    
        if(q[3]==r[i])
            t=i
        if(t==1)
            g=x1
        end
        if(t==2)
            g=x2
        end
        if(t==3)
            g=x3
        end
        if(t==4)
            g=x4
        end
    end
    
        if(q[4]==r[i])
        t=i
        if(t==1)
            w=x1
        end
        if(t==2)
            w=x2
        end
        if(t==3)
            w=x3
        end
        if(t==4)
            w=x4
        end
    end
 
end

println(ff(b), "    ", ff(g), "    ", ff(w))

while((abs(ff(b)-ff(g))>e) &&(abs(ff(g)-ff(w))>e) && n<100)
    n+=1

    x1=b
    x2=w
    x3=g
if(ff(x1)<ff(x3)<ff(x2))
    b=x1
    g=x3
    w=x2
end
if(ff(x1)<ff(x2)<ff(x3))
    b=x1
    g=x2
    w=x3
end
if(ff(x2)<ff(x1)<ff(x3))
    b=x2
    g=x1
    w=x3
end
if(ff(x2)<ff(x3)<ff(x1))
    b=x2
    g=x3
    w=x1
end
if(ff(x3)<ff(x1)<ff(x2))
    b=x3
    g=x1
    w=x2
end
if(ff(x3)<ff(x2)<ff(x1))
    b=x3
    g=x2
    w=x1
end

mid=(b+g)/2
xr=mid+a*(mid-w)

if(ff(xr)<ff(b))
    xe=(1-y)*mid+y*xr
    
        if(ff(xe)<ff(xr))
        w=xe
   continue
   
        else
            
        w=xr
    continue
    end
end
    
    if(ff(b)<ff(xr)<ff(g))
    w=xr
    continue
    end
    
    if(ff(g)<ff(xr)<ff(w))
    xr1=xr
    xr=w
    w=xr1
end
if(ff(xr)>ff(w))
    
end
xs=B*w+(1-B)*mid
if(ff(xs)<ff(w))
    w=xs
     continue
end
if(ff(xs)>ff(w))
        g=b+(g-b)/2
        w=b+(w-b)/2
   continue
end
end

println(" ",b," ",g," ",w, "\nkol-vo iteraciu: ", n)
```
